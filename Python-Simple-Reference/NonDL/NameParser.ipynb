```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import string
```

### For Boys

```python
from selenium import webdriver
browser=webdriver.Firefox(executable_path='/Users/abhinav.sunderrajan/Downloads/geckodriver')
names_list=[]
meanings_list=[]
```

```python
for char in string.ascii_uppercase:
    website='http://tamilcube.com/babynames/sanskrit-baby-names.aspx?baby=g&term={char}'.format(char=char)
    i=2
    browser.get(website)
    clickedOnce=False
    while True:
        soup = BeautifulSoup(browser.page_source)
        names=soup.findAll('span',id=lambda x: x and x.startswith('ListView1_Label1_'))
        meanings=soup.findAll('span',id=lambda x: x and x.startswith('ListView1_Label2_'))
        for name in names:
            names_list.append(name.find(text=True))
        for meaning in meanings:
            meanings_list.append(meaning.find(text=True))
        
        linkElem =browser.find_element_by_id('DataPager1')
        try:
            child= linkElem.find_element_by_link_text(str(i))
            child.click()
        except:
            try:
                if clickedOnce==False:
                    child= linkElem.find_element_by_link_text("...")
                    child.click()
                    clickedOnce=True
                else:
                    print("finished char {char}".format(char=char))
                    break 
            except:
                print("finished char {char}".format(char=char))
                break
        print(char,i)
        i=i+1
    #break
        
browser.close()       
        
    
   

```

```python
file = open("/Users/abhinav.sunderrajan/Desktop/girls_name.csv","w") 
for name, meaning in zip(names_list,meanings_list):
    if name==None or meaning==None:
        continue
    file.write(name+"\t"+meaning+"\n")
    
file.close() 
    
```
