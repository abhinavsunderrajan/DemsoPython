```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import json
os.chdir("/home/abhinav.sunderrajan/ParseReplay/results/traffic_lights/")
%matplotlib inline 
```

### Load data
I also convert hours into local sg time.

```python
#some constants I'll be using in his notebook
date="2019-01-08"
phase="ETT"
```

```python
df=pd.read_csv(f"results_rtt_and_historic{date}with_intersections.csv",sep="\t")
df["hour_of_day"]=(df["hour_of_day"]+8)%24
df["distance_bucket"]=df.edt/1000.0
df.distance_bucket=df.distance_bucket.astype(int)

mask=None
if phase=="ETT":
    mask=(df.phase=="ETT") & (df.distance_bucket<=30) & (df.edt/df.att<=25.0)
else:
    mask=(df.phase=="ETA") & (df.distance_bucket<=5) & (df.edt/df.att<=25.0)

df=df.loc[mask]
df["bias_rtt"]=df.att-df.rtt_expected
df["bias_historic"]=df.att- df.historic_expected
df.head()
```

### Add traffic light information

```python
intersection_delay={}
intersection_delay["motorway"]=30
intersection_delay["residential"]=20
intersection_delay["primary"]=25
intersection_delay["secondary"]=20
intersection_delay["service"]=20
intersection_delay["tertiary"]=20
intersection_delay["trunk"]=35
intersection_delay["unclassified"]=20
```

```python
peak_hours=[8,9,17,18]
def count_num_intersections(row):
    data=json.loads(row["intersections"])
    delay=0
    for key, val in data.items():
        delay=delay+intersection_delay[val.replace("_link","")]
    
    if row["hour_of_day"] not in peak_hours:
        delay=delay//3
 
    retval={}
    retval["num_of_intersections"]=len(data)
    retval["intersection_delay"]=delay
    return pd.Series(retval)

ret=df.apply(count_num_intersections, axis=1)
```

```python
df["num_of_intersections"]=ret["num_of_intersections"]
df["intersection_delay"]=ret["intersection_delay"]
print(df[["num_of_intersections","edt","bias_rtt","bias_historic"]].corr())
df.head()
```

```python
def get_metrics(group):
    metrics={}
    y_true=group['att']
    y_historic=group['historic_expected']
    y_rtt=group['rtt_expected']
    
    ape_ett=np.abs(y_true-y_historic)/y_true
    ape_ett=ape_ett[ape_ett< np.percentile(ape_ett,99)]
    metrics["MAPE_historic"]=ape_ett.mean()
    ape_ett=np.abs(y_true-y_rtt)/y_true
    ape_ett=ape_ett[ape_ett< np.percentile(ape_ett,99)]
    metrics["MAPE_RTT"]=ape_ett.mean()
    
    
    sq_error_ett = (y_true-y_historic) ** 2
    sq_error_ett=sq_error_ett[sq_error_ett< np.percentile(sq_error_ett,99)]
    metrics["SQUARE_ERROR_historic"]=np.sqrt(sq_error_ett.mean())
    sq_error_ett = (y_true-y_rtt) ** 2
    sq_error_ett=sq_error_ett[sq_error_ett< np.percentile(sq_error_ett,99)]
    metrics["SQUARE_ERROR_RTT"]=np.sqrt(sq_error_ett.mean())
    
    abs_error_ett = abs(y_true-y_historic)
    abs_error_ett=abs_error_ett[abs_error_ett< np.percentile(abs_error_ett,99)]
    metrics["ABS_ERROR_historic"]=abs_error_ett.mean()
    abs_error_ett = abs(y_true-y_rtt)
    abs_error_ett=abs_error_ett[abs_error_ett< np.percentile(abs_error_ett,99)]
    metrics["ABS_ERROR_RTT"]=abs_error_ett.mean()
    
    ett_bias=y_true-y_historic
    ett_bias=ett_bias[ett_bias< np.percentile(ett_bias,99)]
    metrics["bias_historic"]=ett_bias.mean()
    ett_bias=y_true-y_rtt
    ett_bias=ett_bias[ett_bias< np.percentile(ett_bias,99)]
    metrics["bias_RTT"]=ett_bias.mean()

    return pd.Series(metrics)
```

### Metrics before applying intersection delay

```python
result=df.groupby(["hour_of_day","distance_bucket","num_of_intersections"]).apply(get_metrics).reset_index()
print(result.shape)
result.dropna(inplace=True)
print(result.shape)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2,figsize=(18,18))

fig.suptitle(date+" "+phase+" no intersection penalty", fontsize=16)
#by hour
by_hour=result.groupby("hour_of_day")["MAPE_RTT","MAPE_historic","bias_RTT","bias_historic"].mean().reset_index()
by_hour.plot(x="hour_of_day",y="MAPE_RTT",kind="line",ax=axes[0,0])
by_hour.plot(x="hour_of_day",y="MAPE_historic",kind="line",ax=axes[0,0])

by_hour.plot(x="hour_of_day",y="bias_RTT",kind="line",ax=axes[0,1])
by_hour.plot(x="hour_of_day",y="bias_historic",kind="line",ax=axes[0,1])

#by distance

by_dist=result.groupby("distance_bucket")["MAPE_RTT","MAPE_historic","bias_RTT","bias_historic"].mean().reset_index()
by_dist.plot(x="distance_bucket",y="MAPE_RTT",kind="line",ax=axes[1,0])
by_dist.plot(x="distance_bucket",y="MAPE_historic",kind="line",ax=axes[1,0])

by_dist.plot(x="distance_bucket",y="bias_RTT",kind="line",ax=axes[1,1])
by_dist.plot(x="distance_bucket",y="bias_historic",kind="line",ax=axes[1,1])


#by intersections

by_intersections=result.groupby("num_of_intersections")["MAPE_RTT","MAPE_historic","bias_RTT","bias_historic"].mean().reset_index()
by_intersections.plot(x="num_of_intersections",y="MAPE_RTT",kind="line",ax=axes[2,0])
by_intersections.plot(x="num_of_intersections",y="MAPE_historic",kind="line",ax=axes[2,0])

by_intersections.plot(x="num_of_intersections",y="bias_RTT",kind="line",ax=axes[2,1])
by_intersections.plot(x="num_of_intersections",y="bias_historic",kind="line",ax=axes[2,1])


```

```python
pain=300.0
if phase=="ETA":
    pain=120.0


painful_historic=df.loc[np.abs(df.historic_expected-df.att)>=pain].shape[0]*100/df.shape[0]
painful_rtt=df.loc[np.abs(df.rtt_expected-df.att)>=pain].shape[0]*100/df.shape[0]

print("Painful journey historic:", painful_historic)
print("Painful journey real time:", painful_rtt)
```
