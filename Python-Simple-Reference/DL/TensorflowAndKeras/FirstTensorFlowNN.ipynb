```python
import tensorflow as tf
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
tf.set_random_seed(42)
%matplotlib notebook
```

```python
n_input=2
n_dense=4
```

```python
x=tf.placeholder(tf.float32,shape=[None,n_input])
w=tf.Variable(tf.random_normal([n_input,n_dense]))
b=tf.Variable(tf.zeros(n_dense))
```

```python
#simple sigmoid functions
z=tf.add(tf.matmul(x,w),b)
a=tf.sigmoid(z)
```

```python
#initialze the variables
initializer_operation=tf.global_variables_initializer()
with tf.Session() as session:
    session.run(initializer_operation)
    layer_output=session.run(a,feed_dict={x:np.random.random([1,n_input])})
```

```python
plt.hist(np.transpose(layer_output))
```
